7) Classify movies according to rating (Exellent, Good, Average, Horrible)

```python
df1=df[['Title','Rating']]
def grade(rating):
    if rating>=8.0:
        return 'Excellent'
    elif rating>=7 and rating<8.0:
        return 'Good'
    elif rating>=5 and rating<7.0:
        return 'Average'
    else:
        return 'Horibble'
df1['Grade']=df1.Rating.apply(grade)
df1
```

```python
df2=df1.groupby('Grade').size().reset_index(name='counts')
df2
```

```python
l=['Average', 'Excellent', 'Good', 'Horrible']
c=['orange','yellow','pink','green']
plt.pie(df2.counts,labels=l,startangle=0,explode=[0.1,0,0,0],shadow=True,autopct='%.2f%%',colors=c)
```

9) In how many movies the actors - Mark Wahlberg, Christian Bale, Scarlett Johansson, Ben Kingsley worked ?

```python
df_act=df[['Title','Rating']]
def fa(actnames):
    actnmlist=[]
    actnmlist=actnames.split(',')
    return actnmlist[0].strip()
def sa(actnames):
    actnmlist=[]
    actnmlist=actnames.split(',')
    return actnmlist[1].strip()
def ta(actnames):
    actnmlist=[]
    actnmlist=actnames.split(',')
    return actnmlist[2].strip()
def foura(actnames):
    actnmlist=[]
    actnmlist=actnames.split(',')
    return actnmlist[len(actnmlist)-1].strip()
df_act['Actor1']=df.Actors.apply(fa).astype('string')
df_act['Actor2']=df.Actors.apply(sa).astype('string')
df_act['Actor3']=df.Actors.apply(ta).astype('string')
df_act['Actor4']=df.Actors.apply(foura).astype('string')
df_act
```

```python
mw=df_act[(df_act.Actor1=='Mark Wahlberg')|(df_act.Actor2=='Mark Wahlberg')|(df_act.Actor3=='Mark Wahlberg')|(df_act.Actor4=='Mark Wahlberg')].shape[0]
cb=df_act[(df_act.Actor1=='Christian Bale')|(df_act.Actor2=='Christian Bale')|(df_act.Actor3=='Christian Bale')|(df_act.Actor4=='Christian Bale')].shape[0]
sj=df_act[(df_act.Actor1=='Scarlett Johansson')|(df_act.Actor2=='Scarlett Johansson')|(df_act.Actor3=='Scarlett Johansson')|(df_act.Actor4=='Scarlett Johansson')].shape[0]
bk=df_act[(df_act.Actor1=='Ben Kingsley')|(df_act.Actor2=='Ben Kingsley')|(df_act.Actor3=='Ben Kingsley')|(df_act.Actor4=='Ben Kingsley')].shape[0]
l=[mw,cb,sj,bk]
aname=['Mark Wahlberg', 'Christian Bale', 'Scarlett Johansson', 'Ben Kingsley']
sns.barplot(x=aname,y=l,palette='prism')
plt.ylabel('no. of movies')
plt.xlabel('actors')
```

1) Predict revenue based on no. of votes.

```python
plt.figure(figsize=(10,5))
plt.title('Revenue vs Votes')
plt.xlabel('Votes')
plt.ylabel('Revenue')
sns.scatterplot(x=df.Votes, y=df['Revenue (Millions)'], palette='prism')
```

From plot, we understand that it is inward heteroskadesticity. Hence we will use simple linear regression.

```python
vote=np.array(df.Votes).reshape(-1,1)
revenue=np.array(df['Revenue (Millions)']).reshape(-1,1)

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(vote, revenue, test_size=0.2, random_state=1)

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain, ytrain)
ypred=lr.predict(xtest)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae=mean_absolute_error(ytest, ypred)
mse=mean_squared_error(ytest, ypred)
r2 = r2_score(ytest, ypred)
print('mean absolute error =',mae)
print('mean square error =',mse)
print('r square =',r2)
```

```python
print('predicted value is',lr.predict([[757074]]))
print('Actual value is 333.13')
```

```python
ypred2=lr.predict(vote)
ypred2=ypred2.flatten()
sns.lineplot(x=df.Votes, y=ypred2, palette='prism')
sns.scatterplot(x=df.Votes, y=df['Revenue (Millions)'], color='red')
plt.title('predicted line')
plt.show()
```

Checking whether using polynomial regression improves r square.

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=4, include_bias=False)
arr1=pf.fit_transform(xtrain)
arr2=pf.fit_transform(xtest)
lr.fit(arr1, ytrain)
ypred3=lr.predict(arr2)
r2 = r2_score(ytest, ypred3)
r2
```

```python
print('predicted value  for 757074 is ',lr.predict([[757074, 757074**2, 757074**3, 757074**4]]))
print('actual value is 333.13')
```

2) Predict metascore based on IMDB rating.

```python
vote=np.array(df.Rating).reshape(-1,1)
revenue=np.array(df['Metascore']).reshape(-1,1)

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(vote, revenue, test_size=0.2, random_state=1)

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain, ytrain)
ypred=lr.predict(xtest)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae=mean_absolute_error(ytest, ypred)
mse=mean_squared_error(ytest, ypred)
r2 = r2_score(ytest, ypred)
print('mean absolute error =',mae)
print('mean square error =',mse)
print('r square =',r2)
```

```python
ypred2=lr.predict(vote)
ypred2=ypred2.flatten()
sns.lineplot(x=df.Rating, y=ypred2, palette='prism')
sns.scatterplot(x=df.Rating, y=df['Metascore'], color='red')
plt.title('predicted line')
plt.show()
```

Using polynomial regression and find r square.

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=3, include_bias=False)
arr1=pf.fit_transform(xtrain)
arr2=pf.fit_transform(xtest)
lr.fit(arr1, ytrain)
ypred3=lr.predict(arr2)
r2 = r2_score(ytest, ypred3)
r2
```

```python
arr3=pf.fit_transform(vote)
ypred4=lr.predict(arr3).flatten()
sns.lineplot(x=df.Rating, y=ypred4, palette='prism')
sns.scatterplot(x=df.Rating, y=df['Metascore'], color='red')
plt.title('predicted curve')
plt.show()
```

```python
print('predicted metascore for 8.1 IMDB rating',lr.predict([[8.1,8.1**2,8.1**3]]))
print('Actual metascore is 76')
```
